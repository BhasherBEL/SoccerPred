```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
pd.set_option('precision', 2)
```

```python
data = pd.read_csv('https://datahub.io/sports-data/german-bundesliga/r/season-1819.csv')[['HomeTeam','AwayTeam','FTHG','FTAG']]
data = data.rename(columns={'FTHG': 'HomeGoals', 'FTAG': 'AwayGoals'})
```

```python
teams = list({el: '' for el in data['HomeTeam']}.keys())
```

```python
classement = {}
for team in teams:
    classement[team] = {
        'MJ': 0,
        'G': 0,
        'N': 0,
        'P': 0,
    }
    for i, line in data[data['HomeTeam'] == team].iterrows():
        classement[team]['MJ']+=1
        if line['HomeGoals'] > line['AwayGoals']:
            classement[team]['G']+=1
        elif line['HomeGoals'] < line['AwayGoals']:
            classement[team]['P']+=1
        else:
            classement[team]['N']+=1
    for i, line in data[data['AwayTeam'] == team].iterrows():
        classement[team]['MJ']+=1
        if line['HomeGoals'] < line['AwayGoals']:
            classement[team]['G']+=1
        elif line['HomeGoals'] > line['AwayGoals']:
            classement[team]['P']+=1
        else:
            classement[team]['N']+=1
df_classement = pd.DataFrame(classement).T
df_classement['pts'] = df_classement['G']*3 + df_classement['N']
df_classement['score'] = df_classement['pts']/df_classement['pts'].sum()*len(teams)
df_classement = df_classement.sort_values('pts', ascending=False).reset_index()
```

```python
teams = list(df_classement['index'])
```

```python
df_classement.head()
```

```python
data['coef'] = 0
for i, match in data.iterrows():
    winner = ''
    loser = ''
    if match['HomeGoals'] > match['AwayGoals']:
        winner = match['HomeTeam']
        loser = match['AwayTeam']
    else:
        winner = match['AwayTeam']
        loser = match['HomeTeam']

    coef = df_classement[df_classement['index'] == loser]['score'].values[0]/df_classement[df_classement['index'] == winner]['score'].values[0]
    
    if match['HomeGoals'] == match['AwayGoals']:
        coef = 1
    
    data.loc[i, 'coef'] = coef
    #df_classement.loc[df_classement['index'] == winner, 'pond_pts'] += coef
```

```python
# importing the tools required for the Poisson regression model
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import poisson
```

```python
goal_model_data = pd.concat([data[['HomeTeam','AwayTeam','HomeGoals']].assign(home=1).rename(
            columns={'HomeTeam':'team', 'AwayTeam':'opponent','HomeGoals':'goals'}),
           data[['AwayTeam','HomeTeam','AwayGoals']].assign(home=0).rename(
            columns={'AwayTeam':'team', 'HomeTeam':'opponent','AwayGoals':'goals'})])

poisson_model = smf.glm(formula="goals ~ home + team + opponent", data=goal_model_data, 
                        family=sm.families.Poisson()).fit()
```

```python
poisson_model.summary()
```

```python
def simulate_match(foot_model, homeTeam, awayTeam, max_goals=10):
    home_goals_avg = foot_model.predict(pd.DataFrame(data={'team': homeTeam, 
                                                            'opponent': awayTeam,'home':1},
                                                      index=[1])).values[0]
    away_goals_avg = foot_model.predict(pd.DataFrame(data={'team': awayTeam, 
                                                            'opponent': homeTeam,'home':0},
                                                      index=[1])).values[0]
    team_pred = [[poisson.pmf(i, team_avg) for i in range(0, max_goals+1)] for team_avg in [home_goals_avg, away_goals_avg]]
    df = pd.DataFrame(np.outer(np.array(team_pred[0]), np.array(team_pred[1])))
    df.columns.name = homeTeam + "\\" + awayTeam
    df = df.style.background_gradient(cmap='Blues', axis=None)
    return df
```

```python
p_pred = np.array([np.sum(np.tril(match.data, -1)), np.sum(np.diag(match.data)), np.sum(np.triu(match.data, 1))])
p_unibet = np.array([1/4.4, 1/4.1, 1/1.74])
```

```python
def style_diag(data):
    diag_mask = pd.DataFrame("", index=data.index, columns=data.columns)
    min_axis = min(diag_mask.shape)
    diag_mask.iloc[range(min_axis), range(min_axis)] = 'background-color: black'
    return diag_mask
```

```python
matches = {}
for hometeam in teams:
    matches[hometeam] = {}
    for awayteam in teams:
        match = simulate_match(poisson_model, hometeam, awayteam, max_goals=10)
        matches[hometeam][awayteam] = np.sum(np.tril(match.data, -1))*100
```

```python
df = pd.DataFrame(matches).round(2).style.background_gradient(cmap='Blues', axis=None).apply(style_diag, axis=None)
df.data.columns.name = "Away/Home"
np.fill_diagonal(df.data.values, 0)
df
```
